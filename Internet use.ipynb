# 3.2. Percentage of households with Internet by year and age groupÂ¶

```python
#Create datafram for percentage of households with internet by year and age group
hou_dem = pd.read_excel('internet.xlsx', sheet_name='2', header=2, usecols='B:I',
                     skiprows=[0,1,6,11,14])
                    
hou_dem=hou_dem.dropna()
hou_dem.columns =['age', '2012', '2013', '2014', '2015', '2016', '2017', '2018']


hou_dem
```

```python
#Data types
hou_dem.dtypes
```

```python
#Set age as category
hou_dem['age'] =hou_dem['age'].astype('category')
hou_dem
```

```python
#Data categories
hou_dem.dtypes
```

```python
#Table of data
hou_dem
```

```python
#Drop all not applicable data 
hou_dem=hou_dem.dropna()
hou_dem
```

```python
#Assign new variable for time series
ts = hou_dem.melt(id_vars=['age'],
var_name='year', value_name='population')
ts.head(28)
```

```python
# Create a datetime series parsing the year strings.
# The data are mid-year estimates, so create timestampes
# to represent this.
ts['year'] = ts['year'].astype('str') +'/06/30'
ts['year'] = pd.to_datetime(ts['year'], format='%Y/%m/%d' )
# Assign the year column to the index, creating a DatetimeIndex.
ts.set_index('year', inplace=True)
# Conver the the DatetimeIndex to a PeriodIndex of day frequency.
ts = ts.to_period('D')
ts.head()
```

```python
# Assign all indicator variables to the index.
ts.set_index(['age'], inplace=True, append=True)
ts.sort_index(inplace=True)
ts.head()
```

```python
ts.loc[('2012-06-30', '1 adult aged 65+')].head(5)
```

```python
#Shows the index is now a day
ts.index
```

```python
#Cross section
ts.xs('3+ adults all ages', level='age').head(7)
```

```python
# 2 adults aged 16 to 64
ts.xs(['2012-06-30', '2 adults aged 16 to 64'],
level=['year', 'age']).head()
```

```python
#plot ages to analyse the change in house composition 2 adults aged 16 to 64 with access to the internet
from matplotlib.ticker import EngFormatter

hou_chil = ts.xs('2 adults aged 16 to 64', level='age').unstack()
ax = hou_chil['population'].plot()
ax.yaxis.set_major_formatter(EngFormatter(places=1))

plt.title('Internet access: 2 adults aged 16 to 64')

plt.ylabel('%')
plt.show()
```

```python
##plot ages to analyse the change in house composition of internet use of 3+ adults of all ages 
from matplotlib.ticker import EngFormatter
hou_chil = ts.xs('3+ adults all ages', level='age').unstack()
ax = hou_chil['population'].plot()
ax.yaxis.set_major_formatter(EngFormatter(places=1))


plt.title('Internet access: 3+ adults all ages')

plt.ylabel('%')

plt.show()
```

```python
#plot ages to analyse the change in house composition of internet use of 1 adult aged 65+
from matplotlib.ticker import EngFormatter
hou_chil = ts.xs('1 adult aged 65+',level='age').unstack()

ax = hou_chil['population'].plot()
ax.yaxis.set_major_formatter(EngFormatter(places=1))

plt.title('Internet access: 1 adult aged 65+')

plt.ylabel('%')

plt.show()
```

# 3.6 Internet activities by age

```python
#Create a new dataframe created from Sheet 9
act = pd.read_excel('internet.xlsx', sheet_name='10', header=6, usecols='B:H',
                     skiprows=[2,8,10,28], nrows=14)

#Rename columns

act.rename(columns={'Unnamed: 0': 'int_act', 'Unnamed: 1': '16-24', 'Unnamed: 2': '25-34', 'Unnamed: 3': '35-44','Unnamed: 4': '45-54', 'Unnamed: 5': '55-64', 'Unnamed: 6': '65+'}, inplace=True)

#Set the category types
act['int_act'] = act['int_act'].astype('category')
act['16-24'] = act['16-24'].astype('float64')
act['25-34'] = act['25-34'].astype('float64')
act['35-44'] = act['35-44'].astype('float64')
act['45-54'] = act['45-54'].astype('float64')
act['55-64'] = act['55-64'].astype('float64')
act['65+'] = act['65+'].astype('float64')
#Set datatypes

#Index the finding_info column
act.rename(index={0: 'Finding_info'})
 

#Index int_act
act.set_index(['int_act'], inplace=True)

act.rename(index={'Finding information about goods or services':'Finding info', 
                 'Watching videos on Youtube or similar':'Watching Youtube',
                 'Watching internet streamed TV or catch-up':'Streaming tv/catchup',
                  'Looking for health-related information (eg injury, disease, nutrition, improving health etc)':'Looking for health info', 
                   'Watching videos on demand from commercial services (eg Netflix)':'Videos on demand',
                 'Making video or voice calls over the internet (eg via Skype or Facetime)':'Making video calls (Skype)',
                 'Playing or downloading games':'Play/Download games',
                 'Selling goods or services over the internet':'Selling',
                 'Making an appointment with a doctor or other healthcare practitioner via a website or app':'Making appointments:Dr etc'},
              inplace=True)



act=act.dropna()
act.iloc[:, 0:5].head(4)

act
```
