```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

#defining and reading csv file from file.
df = pd.read_csv('/Users/coreenjohn/Downloads/corr.csv', index_col=0)

# Pandas.plot methods can simplify code even further, but requires
# OO-style. Pandas.plot methods return axes objects that we can
# manipualte in the usual way to customise a plot. If we want to draw
# on existing axes objects we pass the axes object to the plot method
# using the `ax` parameter.

#Create a figure object of sixe 9x6 inches, and an array of axes
#objects.  The shape of the axes array is (2,3): 2 rows, 3 columns.
fig, ax = plt.subplots(2,3, figsize = (9,6))

#Plot target on the first axes obj, ect.
df.hist('target', ax = ax[0,0])

#Plot target vs a on the second axes object
df.plot.scatter('target', 'a', ax=ax[0,1])

#Plot target vs b on the third axes object
df.plot.scatter('target', 'b', ax=ax[0,2])

#Plot target vs c on the fourth axes object
df.plot.scatter('target', 'c', ax=ax[1,0])

#Plot target vs d on the fifth axes object.
df.plot.scatter('target', 'd', ax=ax[1,1])

#Plot target vs e on the six axes object.
df.plot.scatter('target', 'e', ax=ax[1,2])

#Increase the space between subplots
fig.subplots_adjust(wspace=0.4, hspace=0.4)

#Show the plot
plt.show()
```
