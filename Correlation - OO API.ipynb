```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

#defining and reading csv file from file.
df = pd.read_csv('/Users/coreenjohn/Downloads/corr.csv', index_col=0)

#The OO API is a hierarchically structured set of classes and
# methods. It requires code to be more explicit, which becomes more
# useful as visualisations become more complex. This example is
# arguably simpler in pyplot style.

# In OO-style, we are responsible for managing our own state, i.e. we
# need to define figure and axes variables explicitly.

# Create a figure object of size 9x6 inches, and an array of axes
# objects. The shape of the axes array is (2, 3): 2 rows, 3 columns.
fig, ax=plt.subplots(2,3, figsize = (9,6))

#Plot target distribution on the first axes object (row 0, col 0).
ax[0,0].hist(df['target'])
ax[0,0].set_title('target distrubution') # NB the axis method is 'set_title()'.

#Plot target vs a on the second axes object (row 0, col 1).
ax[0,1].scatter(df['target'], df['a'])
ax[0,1].set_xlabel('target') #NB the axis method is 'set_xlabel()'.
ax[0,1].set_ylabel('a')

#Plot target vs b on the third axes object (row 0, col 2)
ax[0,2].scatter(df['target'], df['b'])
ax[0,2].set_xlabel('target')
ax[0,2].set_ylabel('b')

#Plot target vs c on the fourth axes object (row 1, col 0).
ax[1,0].scatter(df['target'], df['c'])
ax[1,0].set_xlabel('target')
ax[1,0].set_ylabel('c')

#Plot target vs d on the fifth axes object (row 1, col 1).
ax[1,1].scatter(df['target'], df['d'])
ax[1,1].set_xlabel('target')
ax[1,1].set_ylabel('d')

#Plot target vs e on the sixth axes object (row 1, col 2).
ax[1,2].scatter(df['target'], df['e'])
ax[1,2].set_xlabel('target')
ax[1,2].set_ylabel('e')

#Increase the space between subplots.
fig.subplots_adjust(wspace=0.4, hspace=0.4)

#Show the plot.
plt.show()
```
