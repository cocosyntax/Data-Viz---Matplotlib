```python
%matplotlib inline
import numpy as np
import pandas as pd
from pandas.api.types import CategoricalDtype
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns


df = pd.read_csv('/Users/coreenjohn/Downloads/EU Referendum Survey.csv', usecols=[1,2])

#Categorise by particular columns
df.columns =['gender', 'age']

#display the data frame
df
```

```python
#Display the datatypes
df.dtypes
```

```python
#Change the datatype from 'object' to 'category'
df['gender'] =df['gender'].astype('category')
df['age'] =df['age'].astype('category')

#Describe the data types after the change
df.dtypes
```

```python
#Describe the data 
df.describe()
```

```python
table = pd.crosstab(
    df['gender'], 
    df['age'],
    margins=True,   # compute margins for sorting
)
table.sort_values(by='All', ascending=False, inplace=True)
# Drop all row.
table.drop('All', inplace=True)
table
```

```python
# Heatmap alternative.

ax = plt.axes()


ax.set_title('Age and Gender of Respondents')
sns.heatmap(
    table,
    annot=True,
    cbar=False,
)
plt.show()
```
